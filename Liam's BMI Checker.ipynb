```python
import tkinter as tk
from tkinter import messagebox

bmi_menu = ["METRIC", "IMPERIAL"]


def get_bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obesity"


def show_bmi_window(root, system, show_category):
    input_window = tk.Toplevel(root)
    input_window.title(f"{system} BMI Input")
    input_window.geometry("300x280")
    input_window.resizable(False, False)

    weight_var = tk.StringVar()
    height_var = tk.StringVar()
    result_var = tk.StringVar()
    category_var = tk.StringVar()

    unit_weight = "kg" if system == "METRIC" else "lbs"
    unit_height = "m" if system == "METRIC" else "inches"

    def close_input_window():
        input_window.destroy()
        root.deiconify()
        root.lift()

    def calculate():
        try:
            weight = float(weight_var.get())
            height = float(height_var.get())
            if weight <= 0 or height <= 0:
                raise ValueError

            if system == "METRIC":
                bmi = weight / (height ** 2)
            else:
                bmi = (weight / (height ** 2)) * 703

            result_var.set(f"BMI: {bmi:.1f}")
            if show_category:
                category_var.set(get_bmi_category(bmi))
            else:
                category_var.set("")
        except ValueError:
            messagebox.showerror("Invalid input", "Enter positive numbers for weight and height.")

    tk.Label(input_window, text=f"Weight ({unit_weight}):").pack(pady=(12, 0))
    tk.Entry(input_window, textvariable=weight_var).pack(padx=12, fill="x")
    tk.Label(input_window, text=f"Height ({unit_height}):").pack(pady=(10, 0))
    tk.Entry(input_window, textvariable=height_var).pack(padx=12, fill="x")
    tk.Button(input_window, text="Calculate BMI", width=20, command=calculate).pack(pady=12)
    tk.Label(input_window, textvariable=result_var, font=(None, 12, "bold")).pack(pady=(4, 0))
    tk.Label(input_window, textvariable=category_var, font=(None, 11)).pack(pady=(4, 0))
    tk.Button(input_window, text="Close", width=20, command=close_input_window).pack(pady=8)

    input_window.protocol("WM_DELETE_WINDOW", close_input_window)
    input_window.grab_set()
    input_window.focus_set()
    root.withdraw()


def choose_bmi_system():
    root = tk.Tk()
    root.title("Liam's BMI Checker")
    root.geometry("260x180")
    root.resizable(False, False)

    show_category_var = tk.BooleanVar(value=False)

    tk.Label(root, text="Choose BMI unit system:", font=(None, 11)).pack(pady=(12, 8))
    tk.Button(root, text="METRIC", width=20, command=lambda: show_bmi_window(root, "METRIC", show_category_var.get())).pack(pady=4)
    tk.Button(root, text="IMPERIAL", width=20, command=lambda: show_bmi_window(root, "IMPERIAL", show_category_var.get())).pack(pady=4)
    tk.Checkbutton(root, text="Show BMI category", variable=show_category_var).pack(pady=(8, 0))

    root.mainloop()


choose_bmi_system()
```
